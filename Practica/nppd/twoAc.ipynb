```python
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt

file_path = Path('../../cvs/mnnC4I-GRSa5wuCPhjUmvg_8699e4146ad7419d8f5e069340b789f1_Unicorn_Companies.csv')
if not file_path.exists():
    print("File does not exist.")
else:
    print("File exists.")
    df = pd.read_csv(file_path)
    df.head()
```
